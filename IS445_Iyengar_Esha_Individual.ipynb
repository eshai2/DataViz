This notebook contains all the individual visualizations created using the Spotify Dataset obtained from Kaggle. In this notebook, I'd like to explore different relationships.

```python
# importing all necessary packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import squarify
```

```python
# the dataset link: https://www.kaggle.com/datasets/dhruvildave/spotify-charts?resource=download
# size: 991MB
# license: Database: Open Database, Contents: © Original Authors
# number of instance: 26173514
# number of columns: 9


# creating the main dataframe that I will be working off of
df = pd.read_csv('~/Downloads/charts.csv') # change this to your own personal directory
df
```

```python
# filtering to reduce the size of the dataset
X = df.query("date >= '2021-01-01'")
a = X.query("artist == 'Drake'")
b = a.query("region == 'United States'")
c = b.query("title == 'Way 2 Sexy (with Future & Young Thug)'")
drake_song = c.query("chart == 'top200'")
drake_song
```

```python
# create lsits for each of the columns
date = list(drake_song['date'])
rank = list(drake_song['rank'])
trend = list(drake_song['trend'])

fig, ax = plt.subplots()

# looping through to change marker and color based on trend type
for i in range(len(date)):
    if trend[i] == 'SAME_POSITION':
        marker = 'o'
        color = 'C4'
    elif trend[i] == 'MOVE_UP':
        marker = '^'
        color = 'C2'
    elif trend[i] == 'MOVE_DOWN':
        marker = 'v'
        color = 'C3'
    else:
        marker = '+'
        color = 'C5'
    ax.scatter(date[i], rank[i], marker=marker, color=color)
    
    plt.xlabel('Chart Days 2021')
    plt.gca().set_xticklabels([])
    plt.ylabel('Ranking')
plt.title('Way 2 Sexy By Drake Movement on Chart (2021)')
ax.invert_yaxis()

# creating a legend
legend = plt.legend(loc='lower right', handles=[plt.plot([], [], color='C4', linewidth=3)[0],
                                                 plt.plot([], [], color='C2', linewidth=3)[0],
                                                 plt.plot([], [], color='C3', linewidth=3)[0],
                                                 plt.plot([], [], color='C5', linewidth=3)[0]],
                    labels=['Same Position', 'Moved Up', 'Moved Down', 'New Entry'])

plt.show()
```

```python
songs = list(test_US_2017['title'])

song_date = list(test_US_2017['date'])
# song_date

songs_2017_num1 = list(zip(songs, song_date))
# songs_2017_num1
```

```python
def second_item(data):
    return data[1]

songs_2017_num1.sort(key = second_item)
songs_2017_num1
```

```python
num = {}
for i in songs:
    if i in num:
        num[i]+=1
    else:
        num[i]=1
num

song_titles = list(num.keys())
song_count = list(num.values())
```

```python
plt.figure(figsize=(15,5), dpi= 300)
squarify.plot(sizes=song_count, label=song_titles[:8], text_kwargs={'wrap':True}, color = sns.color_palette("magma", 
                                     len(song_count)), alpha = 0.7)
plt.title('Treemap of Number 1 Ranked Songs Based on How long they were on the charts')
plt.axis('off')
plt.show()
```
